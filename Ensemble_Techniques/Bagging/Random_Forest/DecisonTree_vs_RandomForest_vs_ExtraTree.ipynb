# Import libraries and dataset

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import ExtraTreesClassifier
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("F:\machine learning\ML\Ensemble Technique\Random Forest\heart1.csv")
```

```python
df
```

```python
df.isnull().sum()
```

# spliting data

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x_train ,x_test , y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Decision Tree

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)    
```

```python
y_pred_dt_test = dt.predict(x_test)
```

```python
accuracy_score(y_pred_dt_test,y_test)
```

# Hyper parameter tuning for Decision Tree

```python
param_grid = {"max_depth":[3,4,5,6,7,8,9,10],
              "min_samples_split":[2,3,4,5,6,7,8],
              "min_samples_leaf":[1,2,3,4,5,6,7,8],
              "criterion":["gini","entropy"],
              "splitter":["best","random"],
              "random_state":[0,42]}
```

```python
grid = GridSearchCV(dt,param_grid,cv=5,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
b = grid.best_params_
b
```

```python
grid.best_score_
```

```python
dtree = DecisionTreeClassifier(**b)
```

```python
dtree.fit(x_train,y_train)
```

```python
y_pred_train_h = dtree.predict(x_train)
y_pred_test_h = dtree.predict(x_test)
```

```python
print(f"training  {accuracy_score(y_train , y_pred_train_h)}")
print(f"testing   {accuracy_score(y_test , y_pred_test_h)}")
```

# Random Forest 

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
print(f"training  {accuracy_score(y_train , y_pred_rf_train)}")
print(f"testing   {accuracy_score(y_test , y_pred_rf_test)}")
```

# using decison tree hyper parameter

```python
rf2 = RandomForestClassifier(criterion= 'gini',
 max_depth= 6,
 min_samples_leaf= 7,
 min_samples_split= 2,
 random_state= 42,oob_score=True)
```

```python
rf2.fit(x_train,y_train)
```

```python
y_pred_rf_train_h = rf2.predict(x_train)
y_pred_rf_test_h = rf2.predict(x_test)
```

```python
print(f"training  {accuracy_score(y_train , y_pred_rf_train_h)}")
print(f"testing   {accuracy_score(y_test , y_pred_rf_test_h)}")
```

```python
rf2.oob_score
```

# Now doing Hyper parameter tuning for Random Foreast

```python
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees
    'max_depth': [ 10, 20, 30],  # Depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples to split
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples at a leaf node
    'bootstrap': [True, False]}  # Whether to use bootstrapping or not
```

```python
grid2 = GridSearchCV(rf,param_grid,n_jobs=-1,cv=5,verbose=True)
```

```python
grid2.fit(x_train,y_train)
```

```python
best = grid2.best_params_
best
```

```python
grid2.best_score_
```

```python
random = RandomForestClassifier(**best)
```

```python
random.fit(x_train,y_train)
```

```python
best_rf_clf = grid2.best_estimator_
y_pred_test_rf1 = random.predict(x_test)
y_pred_train_rf1 = random.predict(x_train)
```

```python
print("Accuracy Score of Model on Training Data is =>",round(accuracy_score(y_train,y_pred_train_rf1)*100,2),"%")
print("Accuracy Score of Model on Testing Data  is =>",round(accuracy_score(y_test,y_pred_test_rf1)*100,2),"%")
```
