```python
class PrinceMLR:
    def __init__(self):
        self.b=None
        self.m=None

    def fit(self,x_train,y_train):
        x_train = np.insert(x_train,0,1,axis=1) # i have used to just add 1 in front of each row to make caluclation possible 
        betas = np.linalg.inv(np.dot(x_train.T,x_train)).dot(x_train.T).dot(y_train)
        self.b = betas[0] # intercept
        self.m = betas[1:] #mulyiple slopes
        print(f"your Intercept is {self.b} and your slopes are{self.m}")

    def predict(self,x_train):
        pred=np.dot(x_train,self.m)+self.b
        return pred

    def MyR2_score(self,y_test,y_pred):
        L = []
        residual = (y_test-y_pred)
        residual_sq = residual**2
        rss = residual_sq.sum()
        print(rss)
        tss = ((y_test - y_test.mean())**2).sum()
        print(tss)
        r2 = 1 -(rss/tss)
        return f"your r2 score is {r2}"    
```
