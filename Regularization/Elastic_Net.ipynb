# First Linear REG

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error as mse, r2_score
```

```python
d  = load_diabetes()
d
```

```python
df = pd.DataFrame(d['data'],columns=d['feature_names'])
df['Target'] = d['target']
df
```

```python
# i have made this dataframe just to show how data looks 
```

```python
x_train,x_test,y_train,y_test = train_test_split(d['data'],d['target'],test_size=0.2,random_state=42)
```

```python
# x_train.shape
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_test= lr.predict(x_test)
```

```python
print(f"your r2score for linear regession is {r2_score(y_test,y_pred_test)}")
```

```python
print(f"your RMSE for Linear Regression is {np.sqrt(mse(y_test,y_pred_test))}")
```

```python
print(lr.coef_)
```

# Elastic NET

```python
# len(aplha)
```

```python
el = ElasticNet(alpha= 0.005, l1_ratio= 0.12)
```

```python
el.fit(x_train,y_train)
```

```python
print(f"your Elastic net coef are {el.coef_}")
```

```python
y_pred_el= el.predict(x_test)
```

```python
print(f"your elastic net r2 score is {r2_score(y_test,y_pred_el)}")
```

```python
print(np.sqrt(mse(y_test,y_pred_el)))
```

```python
plt.bar(d['feature_names'],el.coef_);
```
